```python
import itertools

x_range = np.linspace(0, 100, 10)
y_range = np.linspace(0, 100, 10)

teores = np.random.randn(100)
teores_2 = teores.reshape(100,1)

coord = []

for x,y in itertools.product(x_range,y_range):
    coord.append([x,y])
```

```python
coordenadas = np.array(coord)

dataset = np.hstack((coordenadas, teores_2))
```

```python
x = dataset[:,0]
y = dataset[:,1]
teores = dataset[:,2]
```

```python
plt.scatter(x,y,c=teores, cmap='jet')
plt.axis('scaled')
plt.colorbar()
plt.show()
```

```python
grid = {'nx':20, 'ny':10, 'ox':0, 'oy':0, 'sx':1, 'sy':2} 
teores = np.random.randn (10*20)
```

```python
grid['nx']
```

```python
def pixelplot(a,grid):
    a = a.reshape(grid['ny'],grid['nx'])
    xmin = grid['ox'] - 0.5* grid['sx']
    xmax = xmin + grid['nx'] * grid['sx']
    ymin = grid['oy'] - 0.5* grid['sy']
    ymax = xmin + grid['ny'] * grid['sy']
    ext = (xmin,xmax,ymin,ymax)    
    plt.imshow(a, extent=ext)
    plt.show()
    
```

```python
pixelplot(teores,grid)
```
