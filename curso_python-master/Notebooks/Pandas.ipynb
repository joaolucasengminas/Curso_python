```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import itertools
```

```python
x_range = np.linspace(0, 100, 10)
y_range = np.linspace(0, 100, 10)

teores = np.random.randn(100)
teores_2 = teores.reshape(100,1)

coord = []

for x,y in itertools.product(x_range,y_range):
    coord.append([x,y])
```

```python
coordenadas = np.array(coord)

dataset = np.hstack((coordenadas, teores_2))
```

```python
df = pd.DataFrame(dataset,columns=['x','y','teores'])
```

```python
def locmap (x,y, var):
    plt.figure(figsize=(8,8))
    plt.scatter(x,y,c=var)
    plt.colorbar()
    plt.title('Mapa de Amostras')
    plt.xlabel('Easting (m)')
    plt.ylabel('Northing (m)')
    plt.axis ('scaled')
    plt.grid()
    plt.show()
```

```python
locmap(jura.Xloc, jura.Yloc,jura.Ni)
```
