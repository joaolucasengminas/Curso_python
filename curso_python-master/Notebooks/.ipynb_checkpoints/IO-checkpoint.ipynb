```python
import numpy as np
import pandas as pd
```

Importing GEOAS

```python
def df_from_geoas (file):
    f = open(file,'r')
    n = 0
    columns = []
    data = []
    for idx, line in enumerate(f):
        if idx == 1:
            n = int(line)
        if  idx>1 and idx<=(n+1):
            line = line.rstrip()
            columns.append(line)
        if idx > n+1:
            arr = np.array(line.split()).astype(float)
            data.append(arr)
    df = pd.DataFrame(data=np.array(data), columns = columns)
    return df
```

Writing to GEOAS

```python
def df_from_geoeas(file):
    f = open(file)
    n = 0
    columns = []
    data = []
    for idx, line in enumerate(f):
        if idx == 1:
            n = int(line)
        if idx > 1 and idx <= (n+1):
            line = line.rstrip()
            columns.append(line)
        if idx > (n+1):
            arr = np.array(line.split()).astype(float)
            data.append(arr)
    
    df = pd.DataFrame(data=np.array(data), columns=columns)
    return df

def  geoeas_from_df(df, file):
    df.replace(float('nan'), -999, inplace=True)
    data_string = ''
    nome = 'df_export\n'
    data_string = data_string + nome
    n = str(len(df.columns)) + '\n'
    data_string = data_string + n
    for coluna in df.columns:
        data_string = data_string + coluna+'\n'
    num_data = df.to_string(header=False, index=False)
    data_string = data_string + num_data
    f = open(file, 'w')
    f.write(data_string)
    f.close()
```
